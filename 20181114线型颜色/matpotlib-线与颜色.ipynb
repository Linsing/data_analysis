```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 绘制line

```python
x =np.linspace(-np.pi,np.pi,256,endpoint=True)
```

```python
c,s = np.cos(x),np.sin(x)
```

```python
plt.figure(1)
```

```python
#help(plt.plot)
```

```python
plt.plot(x,c,color='blue',linewidth=2.0,linestyle='--',label='cos',alpha=0.5) # label 稍安勿躁
```

```python
plt.plot(x,s,'r--',label='sin')
```

```python
plt.title('sin & cos')
```

```python
plt.show()
```
