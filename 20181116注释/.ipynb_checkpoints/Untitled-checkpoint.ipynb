```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.rand(100)

# Plot in different subplots
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(data)

plt.subplot(1, 2, 2)
plt.plot(data)

# plt.subplot(1, 2, 1) # Warning occurs here
plt.plot(data + 1)
```

## https://stackoverflow.com/questions/46933824/matplotlib-adding-an-axes-using-the-same-arguments-as-a-previous-axes

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.rand(100)

# Plot in different subplots
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(data)

ax2.plot(data)

ax1.plot(data+1)

plt.show()
```

```python
plt.figure()
Data = np.random.rand(100)
# keep a reference to the first axis
ax1 = plt.subplot(1,2,1)
ax1.plot(Data)

# and a reference to the second axis
ax2 = plt.subplot(1,2,2)
ax2.plot(Data)
# reuse the first axis
ax1.plot(Data+1)
```

```python
plt.figure()
Data = np.random.rand(100)
# keep a reference to the first axis
ax1 = plt.subplot(1,2,1)
ax1.plot(Data)

# and a reference to the second axis
ax2 = plt.subplot(1,2,2)
ax2.plot(Data)
plt.clf() # clear figure
ax1 = plt.subplot(1,2,1)
# reuse the first axis
ax1.plot(Data+1)
```
