```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 绘制scatter

```python
fig = plt.figure()

ax = fig.add_subplot(331)  #如果有返回值,ax 和plt 就一样了
n=128
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)
T=np.arctan2(Y,X) # 上色
#plt.axes([0.025,0.025,0.95,0.95])

ax.scatter(X,Y,s=75,c=T,alpha=.5) #  s -->size
plt.xlim(-1.5,1.5)#x 的范围
plt.xticks([])
plt.ylim(-1.5,1.5) #y的范围
plt.yticks([])
plt.axis()
plt.title('scatter')
plt.xlabel('x')
plt.ylabel('yy')


fig.add_subplot(332)
n=10
X=np.arange(n)
Y1 = (1-X/float(n)*np.random.uniform(0.5,1.0,n))
Y2 = (1-X/float(n)*np.random.uniform(0.5,1.0,n))
plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')
for x,y in zip(X,Y1):
    plt.text(x+0.4,y+0.05,'%.2f'%y,ha='center',va='bottom') # 给上边添加注释
for x,y in zip(X,Y2):
    plt.text(x+0.4,-y-0.05,'%.2f'%y,ha='center',va='top') # 给下边添加注释
    
fig.add_subplot(333)
# pie
n = 20
Z = np.ones(n)
Z[-1] *= 2
plt.pie(Z,explode=Z*.05,colors=['%f' % (i / float(n)) for i in range(n)],
        labels=['%.2f' % (i / float(n)) for i in range(n)]) # explode 扇形的距离 
plt.gca().set_aspect('equal')

# polar 极坐标图
#fig.add_subplot(334) # 默认画折线
fig.add_subplot(334,polar= True)
n = 20
theta = np.arange(0,2*np.pi,2*np.pi/n)
radii = 10 *np.random.rand(n)
#plt.plot(theta,radii) # plt.polar(theta,radii)
plt.polar(theta,radii)

# 热图 heatmap
fig.add_subplot(335)
from matplotlib import cm  #colormap
data = np.random.rand(3,3)
cmap = cm.Blues  # 蓝色系
map = plt.imshow(data,interpolation='nearest',cmap = cmap,aspect='auto',vmin=0,vmax=1) # interpolation 差值  vmin 最小值 白色

# 3D图
from mpl_toolkits.mplot3d import Axes3D
ax=fig.add_subplot(336,projection='3d')
ax.scatter(1,1,3,s = 100)


# 热图
fig.add_subplot(313)
def f(x,y):
    return (1-x/2 + x**5 +y ** 3)*np.exp(-x**2 - y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
plt.contourf(X,Y,f(X,Y),8,alpha=.75,cmap = plt.cm.hot)

plt.savefig('我的图.png')
plt.show()
```
