```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# scatter 散点图

```python
fig = plt.figure()

#ax = fig.add_subplot(331)  # ax 和plt 就一样
n=128
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)
T=np.arctan2(Y,X) # 上色

print(type(T))
#plt.axes([0.025,0.025,0.95,0.95])
plt.axes([1000,1000,1,1],alpha=.5) # 前两个参数 left bottom ,但是没效果  facecolor='b'
#plt.scatter(X,Y,s=75,c='red',alpha=.5) #  s -->size
# plt.scatter(X,Y,s=75,c='#FF00FF') #   RGB 565   rgb8888 alpha RGB
plt.scatter(X,Y,s=75,c=T,alpha=.5) #   RGB 565   rgb8888 alpha RGB
plt.xlim(-1.5,1.5)#x 的范围
#plt.xticks([])
plt.ylim(-1.5,1.5) #y的范围
plt.yticks([])
#plt.axis()
plt.title('scatter')
plt.xlabel('x')
plt.ylabel('y')
```

# 柱状图

```python
fig.add_subplot(332)
n=10
X=np.arange(n)
Y1 = (1-X/float(n)*np.random.uniform(0.5,1.0,n))
Y2 = (1-X/float(n)*np.random.uniform(0.5,1.0,n))
plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='black')
for x,y in zip(X,Y1):
    plt.text(x+0.1,y,'%.2f'%y,ha='center',va='bottom') # 给上边添加注释
for x,y in zip(X,Y2):
    plt.text(x+0.4,-y-0.05,'%.2f'%y,ha='center',va='top') # 给下边添加注释
```
