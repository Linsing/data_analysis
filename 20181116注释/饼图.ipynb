```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
labels = 'Cats', 'Birds', 'Dogs', 'Pandas'
sizes = [25, 35, 30, 10]
explode = (0, 0, 0.05, 0.08)


fig1, ax1 = plt.subplots()

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%',
        shadow=False, startangle=180)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# Show
plt.show()
```

```python
# pie
n = 20
Z = np.ones(n)
Z[-1] *= 2
plt.pie(Z,explode=Z*.05,colors=['%f' % (i / float(n)) for i in range(n)],
        labels=['%.2f' % (i / float(n)) for i in range(n)]) # explode 扇形的距离 
#plt.gca().set_aspect('equal')
plt.gca().set_aspect(1)
# plt.gca().axis('equal')
#plt.gca().axis('auto')
```

```python
labels = 'Cats', 'Birds', 'Dogs', 'Pandas'
sizes = [25, 35, 30, 10]
explode = (0, 0, 0.05, 0.08)


fig1, ax1 = plt.subplots(2,2) # 如果参数创建多个图,ax1 是array,里边是ax(也就是plt),得通过下标取得相应的轴

n = 20
Z = np.ones(n)
Z[-1] *= 2
ax1[0,1].pie(Z,explode=Z*.05,colors=['%f' % (i / float(n)) for i in range(n)],
        labels=['%.2f' % (i / float(n)) for i in range(n)]) # explode 扇形的距离 
#plt.gca().set_aspect('equal')
ax1[0,1].set_aspect(1)


# fig1, ax1 = plt.subplots(2,2)
ax1[1,1].pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%',
        shadow=False, startangle=180)
ax1[1,1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.



# Show
plt.show()
```
