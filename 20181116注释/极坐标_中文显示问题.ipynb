```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']  # 

# 本质上还是描点
# 角度theta和距离极点的长度radii来确定坐标轴的一个点
fig = plt.figure()
#fig.add_subplot(333,polar = True)
n = 6
# 画出正6边形  思路1 将theta,radii 的值的列表一一写出
# 思路2: 在缺口的未闭合的theta中添加一个值,并据此生成radii的值
# 思路3 : 我在0到2*np.pi+np.pi/3 范围内取值,能取到 2*np.pi这个点
theta = np.arange(0,2*np.pi+np.pi/3,2*np.pi/n)  # 角度

#theta = np.concatenate(([theta,[0]]))
np.random.seed(100)
radii = np.array([1]*7)
#plt.plot(theta,radii*10) # plt.polar(theta,radii)
plt.polar(theta,radii*10)
# plt.xticks([1,2,3,4,5,6])
plt.xticks( np.arange(6), ('魔法伤害', '魔法防御', '移动速度', '物理伤害', '物理防御','攻击速度') )

plt.show()
```
